```python
#  Download the data from the link below using web scrapping:
# https://en.wikipedia.org/wiki/Mount_Everest
# Find the count of Total “Mount” in the text.
# Save the extracted plain text contents in file mount.txt in upper case.
# Print the title of this page.

import requests
from bs4 import BeautifulSoup

# Download the data from the link
url = 'https://en.wikipedia.org/wiki/Mount_Everest'
response = requests.get(url)

# Parse the HTML content
soup = BeautifulSoup(response.content, 'html.parser')

# Extract the plain text content
text = soup.get_text()

# Save the extracted plain text contents in file mount.txt in upper case
with open('mount.txt', 'w') as file:
  file.write(text.upper())

# Print the title of this page
print(soup.title.text)

# Count the number of times "Mount" appears in the text
count = text.count('Mount')
print(f"The word 'Mount' appears {count} times in the text.")
```

```python
#Read the following file
# https://www.london.ac.uk/sites/default/files/study-guides/introduction-to-
# natural-language-processing.pdf
# Print only page number 12 and 13 from it.
# Save this contents in a file nlp.txt
# !pip install PyPDF2
import requests
from bs4 import BeautifulSoup
import PyPDF2

# Download the PDF file
url = 'https://www.london.ac.uk/sites/default/files/study-guides/introduction-to-natural-language-processing.pdf'
response = requests.get(url)

# Open the downloaded PDF file
with open('nlp.pdf', 'wb') as f:
  f.write(response.content)

# Read the PDF file
pdf_file = open('nlp.pdf', 'rb')
pdf_reader = PyPDF2.PdfReader(pdf_file)

# Extract text from pages 12 and 13
# Use the updated method to access pages
page_12_text = pdf_reader.pages[11].extract_text() # Access page 12 (index 11) using pages attribute
page_13_text = pdf_reader.pages[12].extract_text() # Access page 13 (index 12) using pages attribute

# Print the extracted text
print(page_12_text)
print(page_13_text)

# Save the extracted text to a file
with open('nlp.txt', 'w') as f:
  f.write(page_12_text + '\n' + page_13_text)

# Close the PDF file
pdf_file.close()

```

```python
# Read the file from the link givcen below:
# https://icml.cc/Conferences/2013/wp-content/uploads/2013/06/Machine-
# Learning-and-Natural-Language-Processing.docx
# Read and display first 5 paragraphs from it.
# Count total ‘The’ present in the text.
# !pip install python-docx
import requests
from docx import Document

url = 'https://icml.cc/Conferences/2013/wp-content/uploads/2013/06/Machine-Learning-and-Natural-Language-Processing.docx'
response = requests.get(url)

with open('mlnlp.docx', 'wb') as f:
  f.write(response.content)

document = Document('mlnlp.docx')

# Read first 5 paragraphs
for paragraph in document.paragraphs[:5]:
  print(paragraph.text)

# Count total 'The'
text = ' '.join(paragraph.text for paragraph in document.paragraphs)
count = text.count('The')
print(f"\nThe word 'The' appears {count} times in the text.")
```
