```python
from transformers import pipeline
```

```python
classifier = pipeline("text-classification")
```

```python
import pandas as pd
text =  'It was possible for him easily .So he did it .'
result  =  classifier(text)
df =  pd.DataFrame(result)
df
```

```python
text =  'I know that it is   possible for you .'
result  =  classifier(text)
df =  pd.DataFrame(result)
df
```

```python
text =  'I always lied.'
result  =  classifier(text)
df =  pd.DataFrame(result)
df
```

Question Answering

```python
qa =  pipeline ('question-answering')
```

```python
text ="Mumbai is a capital of Maharashtra"
q =  'What is Capital of Maharashtra'
```

```python
a  =qa(question=q , context=text)
```

```python
a['answer']
```

```python
text ="my first name Gaurav. "
q =  'what my first name'
a  =qa(question=q , context=text)
a['answer']
```

##converation

```python
from transformers import Conversation
converse =  pipeline('conversational')
```

```python
conversation = Conversation("Hello ,How are you?")
response =  converse([conversation])
print(response)
```

```python
response[1]['content']
```

# Visual Question answering

```python
from PIL import Image
```

```python
vqa_pipeline =  pipeline('visual-question-answering')
```

```python
image  =   Image.open('German-Shepherd-dog-Alsatian.webp')
```

```python
question  =  'who is there in the image'
```

```python
vqa_pipeline(image,question,top_k =  3)
```

```python

question  = 'what is the color of the dog'
vqa_pipeline(image,question,top_k =  3)
```

```python

question  = 'what is the color of grass'
vqa_pipeline(image,question,top_k =  3)
```

```python

question  = 'where is dog setting'
vqa_pipeline(image,question,top_k =  3)
```

```python
question  = 'whats in picture'
vqa_pipeline(image,question,top_k =  3)
```
