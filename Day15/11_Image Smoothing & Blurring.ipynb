### Averaging Kernel

```python
import cv2
import numpy as np

image = cv2.imread('../images/Lenna.png')

kernel = np.ones((5, 5), np.float32) / 25 # Dividing by 25 results to Averaging Filter

# filter2D (src, dst, ddepth, kernel) = dst is O/P image, ddpeth = Dept of O/P image, -1 will give O/P == I/P image, kernel = 2D matrix for convolution
smooth_image = cv2.filter2D(image, -1, kernel)

cv2.imshow('Original Image', image)
cv2.imshow('Smoothening of Image', smooth_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
new_image = np.hstack((image, smooth_image))
```

```python
cv2.imshow('Both Images', new_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

### Gaussian Kernel

```python
import cv2
import numpy as np

image = cv2.imread('../images/Lenna.png')

kernel_size = 7 # Adjust as needed

sigma = 0.35 * ((kernel_size - 1) * 0.5 - 1) + 0.9 # Gaussian Standard Deviation; the formula is randomly generated

kernel = cv2.getGaussianKernel(kernel_size, sigma)

smooth_image = cv2.filter2D(image, -1, kernel)

cv2.imshow('Original Image', image)
cv2.imshow('Smoothening of Image (Gaussian Kernel)', smooth_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

### Median Filter

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

image = cv2.imread('../images/Lenna.png',0)

kernel_size = 15 # Always in odd size

median = cv2.medianBlur(image, kernel_size)

plt.figure(figsize=(10,6))

plt.subplot(121), plt.imshow(image, cmap='gray'), plt.title('Original Image (in grayscale)')
plt.subplot(122), plt.imshow(median, cmap='gray'), plt.title('Median Filtered Image')
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

original_image = cv2.imread('../images/Lenna.png',0)

kernel_size = 25 # Always in odd size

median = cv2.medianBlur(image[100:300,100:300], kernel_size) # image[row_label, column_label]

image[100:300,100:300] = median

plt.figure(figsize=(10,6))

plt.subplot(121), plt.imshow(original_image, cmap='gray'), plt.title('Original Image (in grayscale)')
plt.subplot(122), plt.imshow(image, cmap='gray'), plt.title('Median Filtered Image')
plt.show()
```
