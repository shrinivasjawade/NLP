```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

image = cv2.imread('../images/Lenna.png',0)

binary_image = image > 127 # False(0/0) is black and True(1/255) is white (based on original Image)

plt.figure(figsize=(10,6))

plt.subplot(121), plt.imshow(image, cmap='gray'), plt.title('Original Image (in grayscale)')
plt.subplot(122), plt.imshow(binary_image, cmap='gray'), plt.title('Binary Image')
plt.show()
```

### Inverse Thresholding

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

image = cv2.imread('../images/Lenna.png',0)

binary_image = image < 127 # True(1/255) is black and False(0/0) is white (based on original Image)

plt.figure(figsize=(10,6))

plt.subplot(121), plt.imshow(image, cmap='gray'), plt.title('Original Image (in grayscale)')
plt.subplot(122), plt.imshow(binary_image, cmap='gray'), plt.title('Inverse Binary Image')
plt.show()
```

```python
binary_image
```

```python
binary_image = binary_image.astype(np.uint8) * 255 # Grayscaling
binary_image
# cv2.imshow('Original Image', image)
# cv2.imshow('Image Thresholding', binary_image)
# cv2.waitKey(0)
# cv2.destroyAllWindows()
```

```python
# cv2.threshold(src,dst, thresh, maxval, type) 
# src: input array (multiple-channel, 8-bit or 32-bit floating point).
# dst: output array of the same size and type and the same number of channels as src.
# thresh: threshold value.
# maxval: maximum value to use with the THRESH_BINARY and THRESH_BINARY_INV thresholding types.
# type: thresholding type.

ret, thresh1 = cv2.threshold(image, 120, 255, cv2.THRESH_BINARY)
ret, thresh2 = cv2.threshold(image, 120, 255, cv2.THRESH_BINARY_INV)

plt.figure(figsize=(10,6))

plt.subplot(1,3,1), plt.imshow(image, cmap='gray'), plt.title('Original Image (in grayscale)')
plt.subplot(1,3,2), plt.imshow(thresh1, cmap='gray'), plt.title('Binary Image')
plt.subplot(1,3,3), plt.imshow(thresh2, cmap='gray'), plt.title('Inverse Binary Image')
plt.show()
```

### Otsu Thersholding

by Noboyuki Otsu , is a method of Adaptive Thresholding

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

gray = cv2.imread('../images/Lenna.png',0)

val, thresh_otsu = cv2.threshold(gray, 0, 255, cv2.THRESH_OTSU) # val = 124

ret, thresh1 = cv2.threshold(image, 120, 255, cv2.THRESH_BINARY)
ret, thresh2 = cv2.threshold(image, 120, 255, cv2.THRESH_BINARY_INV)

plt.figure(figsize=(10,6))

plt.subplot(1,4,1), plt.imshow(gray, cmap='gray'), plt.title('Original Image (in grayscale)')
plt.subplot(1,4,2), plt.imshow(thresh_otsu, cmap='gray'), plt.title('Otsu Image')
plt.subplot(1,4,3), plt.imshow(thresh1, cmap='gray'), plt.title('Binary Image')
plt.subplot(1,4,4), plt.imshow(thresh2, cmap='gray'), plt.title('Inverse Binary Image')
```
