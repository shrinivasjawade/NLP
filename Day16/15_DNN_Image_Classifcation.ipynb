```python
import numpy as np
import matplotlib.pyplot as plt
from keras.models import Sequential 
from  keras.layers import Dense
from keras.utils import to_categorical,plot_model
from keras.datasets import mnist
                            
```

```python
#load the datac
```

```python
(X_train,y_train),(X_test,y_test) =  mnist.load_data()
```

Normalize

```python
X_train =  X_train.reshape(-1,X_train[0].size)
X_test =  X_test.reshape(-1,X_test[0].size)
```

```python
X_train.shape
```

```python
X_test.shape
```

### Preparing the output variable

```python
import seaborn as sns
```

```python
sns.countplot(y  =  y_train);
```

```python
sns.countplot(y  =  y_test);
```

```python
y_train.shape
```

```python
y_train = to_categorical(y_train)
y_test =  to_categorical(y_test)
```

```python
y_train.shape
```

```python
y_test.shape
```

```python
y_train
```

```python
########Build the model 
```

```python
model  =  Sequential()
model.add(Dense(512,input_shape = (784,),activation='relu'))

#hidden Layer
model.add(Dense(128,activation='relu'))

#output layer

model.add(Dense(10,input_shape = (784,),activation='softmax'))
```

```python
model.summary()
```

```python
plot_model(model,show_layer_activations=True,show_layer_names=True,show_shapes=True)
```

```python
##Compile the model 
```

```python
model.compile(optimizer= "adam",loss= "categorical_crossentropy",metrics  =  ['accuracy'])
```

```python
# train the model

model.fit(X_train,y_train,epochs = 10,batch_size =10)
```

```python
##Evaluate on test data 
```

```python
model.evaluate(X_test,y_test,batch_size=1)
```
