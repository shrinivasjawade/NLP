### Importing the Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
from keras.datasets import mnist
from keras.preprocessing.image import ImageDataGenerator
```

### Load the Data

```python
(X_train, y_train), (X_test, y_test) = mnist.load_data()
```

```python
plt.figure(figsize=(16,16)) 
for i in range(1,26):
    plt.subplot(5,5,i)
    plt.imshow(X_train[i], cmap='gray')
    plt.title(y_train[i], fontsize=14)
    plt.axis('off')
```

### ZCA Whitening

```python
X_train = X_train / 255
X_test = X_test / 255
```

```python
# zc whitening is like performing decorrealtion
datagen = ImageDataGenerator(zca_whitening=True, featurewise_center=True)
```

```python
X_train=X_train.reshape(-1,28,28,1)
X_train.shape
```

```python
X_test = X_test.reshape(-1,28,28,1)
X_test.shape
```

```python
datagen.fit(X_train)
```

```python
for X_batch, y_batch in datagen.flow(X_train, y_train, batch_size=25):
    
    plt.figure(figsize=(12, 12))
    for i in range(0, 25):
        plt.subplot(5,5, i+1)
        plt.title(y_batch[i], fontsize=14)
        plt.imshow(X_batch[i].reshape(28, 28), cmap='gray')
        plt.axis('off')
    
    plt.show()
    break
```

### Zooming

```python
datagen = datagen = ImageDataGenerator(zoom_range=0.6)

datagen.fit(X_train)

for X_batch, y_batch in datagen.flow(X_train, y_train, batch_size=25):
    
    plt.figure(figsize=(12, 12))
    for i in range(0, 25):
        plt.subplot(5,5, i+1)
        plt.title(y_batch[i], fontsize=14)
        plt.imshow(X_batch[i].reshape(28, 28), cmap='gray')
        plt.axis('off')
    
    plt.show()
    break
```

### Random Flipping

```python
datagen = ImageDataGenerator(horizontal_flip=True, vertical_flip=True)

datagen.fit(X_train)

for X_batch, y_batch in datagen.flow(X_train, y_train, batch_size=25):
    
    plt.figure(figsize=(12, 12))
    for i in range(0, 25):
        plt.subplot(5,5, i+1)
        plt.title(y_batch[i], fontsize=14)
        plt.imshow(X_batch[i].reshape(28, 28), cmap='gray')
        plt.axis('off')
    
    plt.show()
    break
```

### Random Rotation

```python
datagen = ImageDataGenerator(rotation_range=0.25)

datagen.fit(X_train)

for X_batch, y_batch in datagen.flow(X_train, y_train, batch_size=25):
    
    plt.figure(figsize=(12, 12))
    for i in range(0, 25):
        plt.subplot(5,5, i+1)
        plt.title(y_batch[i], fontsize=14)
        plt.imshow(X_batch[i].reshape(28, 28), cmap='gray')
        plt.axis('off')
    
    plt.show()
    break
```

### Shearing

```python
datagen = ImageDataGenerator(shear_range=0.20)

datagen.fit(X_train)

for X_batch, y_batch in datagen.flow(X_train, y_train, batch_size=25):
    
    plt.figure(figsize=(12, 12))
    for i in range(0, 25):
        plt.subplot(5,5, i+1)
        plt.title(y_batch[i], fontsize=14)
        plt.imshow(X_batch[i].reshape(28, 28), cmap='gray')
        plt.axis('off')
    
    plt.show()
    break
```

### Random Translation / Shift

```python
datagen = ImageDataGenerator(width_shift_range=0.2, height_shift_range=0.2)

datagen.fit(X_train)

for X_batch, y_batch in datagen.flow(X_train, y_train, batch_size=25):
    
    plt.figure(figsize=(12, 12))
    for i in range(0, 25):
        plt.subplot(5,5, i+1)
        plt.title(y_batch[i], fontsize=14)
        plt.imshow(X_batch[i].reshape(28, 28), cmap='gray')
        plt.axis('off')
    
    plt.show()
    break
```
