```python
from urllib.request import urlopen
import be
```

```python
url  =  urlopen('https://en.wikipedia.org/wiki/Raigad_Fort')
data  =  url.read()
print(data)

```

```python
##to extract the text ffrom above data and remove the html tag
import bs4
from bs4 import BeautifulSoup
soup  =  BeautifulSoup(data ,"html.parser") 
```

```python
print(soup)
```

```python
soup.find('title')
```

```python
text   =  soup.get_text()
```

```python
print(text)
```

```python
from nltk.tokenize import word_tokenize
from nltk import FreqDist
```

```python
tokens =   word_tokenize(text)
freq =   FreqDist(tokens)
freq.plot(20)
```

```python
help(FreqDist)
```

```python
from nltk.corpus import stopwords
swords =  stopwords.words('english')


tokens1 = [token for token in tokens if token.lower() not in swords]
tokens2  =  [token.lower() for token in tokens1  if token.isalnum()]
```

```python
tokens =   word_tokenize(text)
freq =   FreqDist(tokens2)
freq.plot(20)
```
