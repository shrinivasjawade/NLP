```python
import re
import string

def cleanResume(resumeText):
    resumeText = re.sub('http\S+\s*', ' ', resumeText) # remove url
    resumeText = re.sub('RT|cc', ' ', resumeText) # remove RT and cc
    resumeText = re.sub('#\S+', ' ', resumeText) # Remove hastag
    resumeText = re.sub('@\S+', ' ', resumeText) # remove mention
    resumeText = re.sub('[%s]'%re.escape(string.punctuation), ' ', resumeText) # REMOE THE PUNCUATION
    resumeText = re.sub(r'[^\x00-\x7f]',r' ', resumeText)
    resumeText = re.sub('\s+', ' ', resumeText)

    return resumeText
```

```python
cleanResume(x[123])
```

```python
cleaned = x.apply(cleanResume)
```

```python
cleaned.shape
```

```python
cleaned
```

##### Tf*IDF Vectorize

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
# Removing the stop words
tfv = TfidfVectorizer(stop_words='english')
```

```python
x_new = tfv.fit_transform(cleaned)
```

```python
x_new
```

```python
tfv.get_feature_names_out()
```

#### Cross Validation

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x_new,y,random_state=0)
```

```python
x_train.shape
```

```python
y_train.shape
```

#### Building Model

```python
from sklearn.neighbors import NearestCentroid

nn =  NearestCentroid()
```

```python
nn.fit(x_train,y_train)
```

#### Performace Evalation

```python
y_pred =  nn.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
accuracy_score(y_pred,y_test)
```

```python
confusion_matrix(y_pred,y_test)
```

```python
print(classification_report(y_pred,y_test))
```
